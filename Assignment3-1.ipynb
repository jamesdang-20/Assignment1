```python
# Coursera Week 3 Assignment
# Problem 1 - Identifying Neighborhoods in Toronto
# Author: James Dang
# Date: May 28, 2020
# Import the libraries used for the assignment

import requests
import pandas as pd
```

```python
wiki = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wikipedia_page = requests.get(wiki)

df_raw = pd.read_html(wikipedia_page.content, header=0)[0]

df_raw.head (20)
```

```python
# Find whether there is a "Not assigned" in Neighbourhood
df_new = df_raw[df_raw.Borough == 'Not assigned']
df_new.head (20)
```

```python
#Find whether there is a not "Not assigned" in Borough
df_new = df_raw[df_raw.Borough != 'Not assigned']
df_new.head (20)
```

```python
#If we have Neighbourhood Not assigned, we change it with the value of Borough
df_new.Neighborhood.replace('Not assigned',df_new.Borough,inplace=True)
df_new.head(20)
```
